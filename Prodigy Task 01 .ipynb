```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('starbucks.csv')
```

```python
df.head()
```

```python
y=df['Cholesterol (mg)']
x=df['Beverage_category']
```

```python
x.isna().sum()
```

```python
y.isna().sum()
```

```python
plt.bar(x,y)
plt.title("Beverage Category and Cholesterol in it")
plt.ylabel('Cholesterol (mg)')
plt.xlabel('Beverage category')
plt.xticks(rotation=45)
```

```python
sns.histplot(y)
```

```python
plt.barh(x,y)
plt.title("Beverage Category and Cholesterol in it")
plt.ylabel('Cholesterol (mg)')
plt.xlabel('Beverage category')
```
